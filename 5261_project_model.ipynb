```python
# 1. 确保日期为 datetime 格式并设置索引（防止 index 对不齐）
train['date'] = pd.to_datetime(train['date'])
test['date'] = pd.to_datetime(test['date'])
train_ts = train.set_index('date')
test_ts = test.set_index('date')

# 2. 构建目标变量（对 boeing 做对数差分）
y_train_log_diff = np.log(train_ts['boeing']).diff().dropna()

# 3. 构建外生变量（去掉 boeing）
X_train = train_ts.drop(columns=['boeing'])
X_test = test_ts.drop(columns=['boeing'])

# 4. 合并外生变量后进行标准化
from sklearn.preprocessing import StandardScaler
X_all = pd.concat([X_train, X_test])
scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X_all),
    index=X_all.index,
    columns=X_all.columns
)

# 5. 拆分标准化后的外生变量，并对齐目标变量
X_train_scaled = X_scaled.loc[y_train_log_diff.index]
X_test_scaled = X_scaled.loc[test_ts.index]
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX

# 1. 创建目标变量（log-diff 的 boeing）
y_train_log_diff = np.log(train_ts['boeing']).diff().dropna()

# 2. 拼接训练和测试的外生变量，做标准化
X_all = pd.concat([X_train, X_test])
scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X_all),
    index=X_all.index,
    columns=X_all.columns
)

# 3. 拆分标准化后的外生变量
X_train_scaled = X_scaled.loc[y_train_log_diff.index]
X_test_scaled = X_scaled.loc[test_ts.index]

# 4. 构建 ARIMAX 模型，使用 SARIMAX 实现
model = SARIMAX(
    y_train_log_diff,
    exog=X_train_scaled,
    order=(1, 0, 1),
    enforce_stationarity=False,
    enforce_invertibility=False
)
results = model.fit(disp=False)

# 5. 展示模型摘要
print(results.summary())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 预测 log-diff 的 boeing
forecast_log_diff = results.predict(
    start=X_test_scaled.index[0],
    end=X_test_scaled.index[-1],
    exog=X_test_scaled
)

# 2. 还原回对数形式（累加）
last_train_log = np.log(train_ts['boeing'].iloc[-1])  # 起点是训练集最后一个 log(boeing)
forecast_log = forecast_log_diff.cumsum() + last_train_log

# 3. 对预测值做指数还原，得到 boeing 股价预测
forecast_price = np.exp(forecast_log)
forecast_price_arimax = forecast_price.copy()


# 4. 实际 boeing 股价（原始 test_ts 中的 boeing 列）
actual_price = test_ts['boeing']

# 5. 绘图：实际 vs 预测
plt.figure(figsize=(12, 6))
plt.plot(actual_price.index, actual_price, label='Actual Boeing Price', linewidth=2)
plt.plot(actual_price.index, forecast_price, label='Predicted Boeing Price (ARIMAX)', linestyle='--', color='orange')
plt.title('ARIMAX Model Forecast vs Actual Boeing Price')
plt.xlabel('Date')
plt.ylabel('Boeing Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.api import VARMAX
from sklearn.preprocessing import StandardScaler
# 内生变量（我们这次不只 boeing，而是联合建模）
Y_cols = ['boeing', 'oil', 'DJI']
X_cols = ['CPI', 'delivery', 'Al']

# 差分后构建内生变量（log-diff or diff）
Y_train = train_ts[Y_cols].diff().dropna()
Y_test = test_ts[Y_cols].diff().dropna()

# 外生变量标准化（X 不差分，只标准化）
X_all = pd.concat([train_ts[X_cols], test_ts[X_cols]])
scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X_all),
    index=X_all.index,
    columns=X_all.columns
)

# 对齐差分后的 Y（不能用前一行差分导致的 NaN）
X_train_scaled = X_scaled.loc[Y_train.index]
X_test_scaled = X_scaled.loc[Y_test.index]
```

```python
# 建立 VARX 模型（注意用 VARMAX，适用于多变量 + exog）
model_varx = VARMAX(Y_train, exog=X_train_scaled, order=(2, 0))  # VAR(2)
results_varx = model_varx.fit(disp=False)
print(results_varx.summary())
```

```python
# 使用 X_test_scaled 做预测
forecast_varx = results_varx.forecast(steps=len(Y_test), exog=X_test_scaled)

# 累加还原差分结果（还原 boeing 的价格）
forecast_boeing_diff = forecast_varx['boeing']
boeing_start = train_ts['boeing'].iloc[-1]
forecast_boeing = boeing_start + forecast_boeing_diff.cumsum()

# 实际 boeing 数据（从 Y_test 中还原）
actual_boeing = test_ts['boeing'].loc[forecast_boeing.index]
```

```python
forecast_price_varx = forecast_boeing.copy()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(actual_boeing.index, actual_boeing, label='Actual Boeing Price', linewidth=2)
plt.plot(actual_boeing.index, forecast_boeing, label='Forecast (VARX)', linestyle='--', color='orange')
plt.title('VARX Model Forecast vs Actual Boeing Price')
plt.xlabel('Date')
plt.ylabel('Boeing Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX
import numpy as np

# 目标变量：对 log(boeing) 做差分（金融时间序列常见做法）
y_train_dyn = np.log(train_ts['boeing']).diff().dropna()
y_test_dyn = np.log(test_ts['boeing']).diff().dropna()

# 外生变量（不差分，但要标准化）
X_cols = ['CPI', 'delivery', 'Al']
X_all = pd.concat([train_ts[X_cols], test_ts[X_cols]])
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_all), index=X_all.index, columns=X_all.columns)

# 对齐外生变量
X_train_scaled = X_scaled.loc[y_train_dyn.index]
X_test_scaled = X_scaled.loc[test_ts.index]
y_test_dyn = np.log(test_ts['boeing']).diff().dropna()
X_test_dyn = X_test_scaled.loc[y_test_dyn.index]
```

```python
model_dyn = SARIMAX(
    y_train_dyn,
    exog=X_train_scaled,
    order=(1, 0, 1),  # 残差建模为 ARMA(1,1)
    enforce_stationarity=False,
    enforce_invertibility=False
)

results_dyn = model_dyn.fit(disp=False)
print(results_dyn.summary())
```

```python
# 预测 log-diff
forecast_logdiff = results_dyn.predict(
    start=X_test_scaled.index[0],
    end=X_test_scaled.index[-1],
    exog=X_test_scaled
)

# 还原 log -> price
last_log = np.log(train_ts['boeing'].iloc[-1])
forecast_log = forecast_logdiff.cumsum() + last_log
forecast_price = np.exp(forecast_log)

# 实际股价
actual_price = test_ts['boeing'].loc[forecast_price.index]
```

```python
forecast_price_dyn = forecast_price.copy() 
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(actual_price.index, actual_price, label='Actual Boeing Price', linewidth=2)
plt.plot(forecast_price.index, forecast_price, label='Forecast (Dynamic Regression)', linestyle='--', color='orange')
plt.title('Dynamic Regression Model Forecast vs Actual Boeing Price')
plt.xlabel('Date')
plt.ylabel('Boeing Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

def compute_metrics(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    rmse = mean_squared_error(actual, predicted, squared=False)
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    return mae, rmse, mape

# 1. ARIMAX
mae_arimax, rmse_arimax, mape_arimax = compute_metrics(actual_price, forecast_price_arimax)

# 2. VARX
mae_varx, rmse_varx, mape_varx = compute_metrics(actual_price, forecast_boeing)

# 3. Dynamic Regression
mae_dyn, rmse_dyn, mape_dyn = compute_metrics(actual_price, forecast_price)

# 汇总展示
metrics_summary = pd.DataFrame({
    'Model': ['ARIMAX', 'VARX', 'Dynamic Regression'],
    'MAE': [mae_arimax, mae_varx, mae_dyn],
    'RMSE': [rmse_arimax, rmse_varx, rmse_dyn],
    'MAPE (%)': [mape_arimax, mape_varx, mape_dyn]
})

metrics_summary.set_index('Model', inplace=True)
metrics_summary
```
