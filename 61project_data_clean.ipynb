```python
delivery = df.iloc[0:301, 14:16]
delivery
```

```python
# Step 1: 修正列名 + 转换日期
delivery.columns = ['date8', 'delivery']
delivery['date8'] = pd.to_datetime(delivery['date8'])

# Step 2: 设置为索引
delivery = delivery.set_index('date8')

# Step 3: 使用 oil_weekly 或 boeing 的日期作为目标周频时间轴
weekly_dates = pd.to_datetime(oil_weekly['date3'])  # 或 boeing.index

# Step 4: 重建并插值为周频
delivery_weekly = pd.DataFrame(index=weekly_dates)
delivery_weekly['delivery'] = delivery.reindex(weekly_dates).interpolate(method='linear')

# Step 5: 可选重设索引为列
delivery_weekly = delivery_weekly.reset_index().rename(columns={'index': 'Date'})

# 查看结果
print(delivery_weekly.head())
print(delivery_weekly.tail())
print(delivery_weekly.shape)  # 应该是 (1311, 2)
```

```python
# Step 1: 确保每个表都 reset_index 并保留原来的日期列名
boeing = boeing.reset_index().rename(columns={'Date': 'date1', 'Close': 'boeing'})
DJI = DJI.reset_index().rename(columns={'Date': 'date2', 'Close': 'DJI'})
dollar = dollar.reset_index().rename(columns={'Date': 'date3', 'Close': 'dollar'})
oil_weekly = oil_weekly.rename(columns={'Date': 'date4'})
gdp_weekly = gdp_weekly.rename(columns={'Date': 'date6'})
cpi_weekly = cpi_weekly.rename(columns={'Date': 'date7'})
delivery_weekly = delivery_weekly.rename(columns={'Date': 'date8'})

# Step 2: 合并为大表格（按列拼接）
weekly_master = pd.concat(
    [boeing, DJI, dollar, oil_weekly, gdp_weekly, cpi_weekly, delivery_weekly],
    axis=1
)

# Step 3: 查看合并结果
print(weekly_master.head())
print(weekly_master.tail())
print(weekly_master.shape)
```

```python
weekly_master.to_csv("weekly_master.csv", index=False, encoding='utf-8-sig')
```

```python
import pandas as pd

# Step 1: 转换日期格式
Al_weekly['date'] = pd.to_datetime(Al_weekly['date'])
Al_monthly['date5'] = pd.to_datetime(Al_monthly['date5'])

# Step 2: 设置日期为索引
Al_weekly = Al_weekly.set_index('date')
Al_monthly = Al_monthly.set_index('date5')

# Step 3: 构建完整的目标周时间轴（从2000/12/1到2025/3/1）
weekly_index = pd.date_range(start='2000-12-01', end='2025-03-01', freq='7D')

# Step 4: 合并数据
# 创建一个空 DataFrame 以合并
Al_combined = pd.DataFrame(index=weekly_index)

# 先用 monthly 的数据补全，interpolate 使其变成周频
Al_combined['AL'] = Al_monthly['AL'].reindex(weekly_index).interpolate(method='linear')

# 再用 weekly 的数据覆盖（优先使用真实的周数据）
Al_combined.update(Al_weekly[['Price']].rename(columns={'Price': 'AL'}))

# Step 5: 重置索引为列
Al_combined = Al_combined.reset_index().rename(columns={'index': 'Date'})

# 查看结果
print(Al_combined.head())
print(Al_combined.tail())
print(Al_combined.shape)

```

```python
# 假设 Al_combined 的日期列名是 'Date'
Al_combined['Date'] = pd.to_datetime(Al_combined['Date']) + pd.Timedelta(days=1)
# 查看结果
print(Al_combined.head())
print(Al_combined.tail())
print(Al_combined.shape)
```

```python
# Step 6: 保存最终数据
Al_combined.to_csv("Al_combined.csv", index=False, encoding='utf-8-sig')
```
